```python
# Select relevant columns including 'has_gas' and 'churn'
consumption = client_df[["cons_12m", "cons_gas_12m", "cons_last_month", "imp_cons", "has_gas", "churn"]]
```

```python
# Convert churn to boolean or int if needed (0 = Retained, 1 = Churned)
consumption['churn'] = consumption['churn'].astype(int)

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distribution in a stacked histogram comparing churned vs retained customers.

    Parameters:
    - dataframe: DataFrame containing the data
    - column: Column name to plot
    - ax: Matplotlib axis object to draw on
    - bins_: Number of bins for the histogram
    """
    # Separate into two groups: retained and churned
    temp = pd.DataFrame({
        "Retention": dataframe[dataframe["churn"] == 0][column],
        "Churn": dataframe[dataframe["churn"] == 1][column]
    })

    # Plot stacked histogram
    temp[["Retention", "Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True, alpha=0.7, edgecolor='black')

    # Set labels and title
    ax.set_xlabel(column.replace('_', ' ').title(), fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title(f'Distribution of {column.replace("_", " ").title()}', fontsize=14)

    # Format x-axis numbers to avoid scientific notation
    ax.ticklabel_format(style='plain', axis='x')
```

```python
# Create subplots
fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(16, 22))

# Plot each consumption metric
plot_distribution(consumption, 'cons_12m', axs[0], bins_=100)
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1], bins_=100)
plot_distribution(consumption, 'cons_last_month', axs[2], bins_=50)
plot_distribution(consumption, 'imp_cons', axs[3], bins_=50)

# Improve spacing between plots
plt.tight_layout()
plt.subplots_adjust(hspace=0.4)  # Increase space between subplots
plt.show()
```

```python
def plot_histogram_by_churn(df, target_variable, ax, bins=100, if_churn=True):
    """
    Plot histogram for a given feature, optionally colored by churn status.

    Parameters:
    - df: DataFrame containing the data
    - target_variable: Column name to plot
    - ax: Matplotlib axis object to draw on
    - bins: Number of bins for the histogram
    - if_churn: Whether to color by 'churn' column (hue)
    """
    if if_churn:
        sns.histplot(data=df, x=target_variable, bins=bins, hue='churn', ax=ax,
                     kde=False, alpha=0.7, edgecolor=None, palette='Set2')
    else:
        sns.histplot(data=df, x=target_variable, bins=bins, ax=ax,
                     kde=False, alpha=0.7, edgecolor=None, color='skyblue')

    # Set title and labels
    ax.set_title(f"Distribution of {target_variable.replace('_', ' ').title()}", fontsize=12)
    ax.set_xlabel(target_variable.replace('_', ' ').title(), fontsize=10)
    ax.set_ylabel("Frequency", fontsize=10)

    # Avoid scientific notation on x-axis
    ax.ticklabel_format(style='plain', axis='x')


# Define the list of price attributes to plot
price_attributes = [
    'price_off_peak_var', 'price_peak_var', 'price_mid_peak_var',
    'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix'
]

# Create subplots: 2 rows x 3 columns to neatly fit all plots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
axes = axes.flatten()

# Loop through each attribute and plot
for i, attr in enumerate(price_attributes):
    plot_histogram_by_churn(df=price_df, target_variable=attr, ax=axes[i], bins=100, if_churn=False)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.suptitle('Price Variable Distributions (Non-Churn Specific)', fontsize=16, y=1.02)
plt.show()
```

# **Hypothesis testing**



*   Calculate average prices over different time periods (full year, last 6 months, last 3 months)

*   Create new features by combining variable + fixed price components



*   Rename columns clearly to avoid confusion when merging later

---

🧠 Why This Matters
This kind of feature engineering helps answer questions like:

* Did customers who left pay higher average prices?
* Was there a sudden price increase in the last 3 months before they left?
* Do customers care more about recent pricing trends than long-term averages?

```python
# Convert 'price_date' to datetime objects
price_df['price_date'] = pd.to_datetime(price_df['price_date']) #This allows you to do date-based filtering like comparing dates or extracting month/year.


# Create mean average data
mean_year = price_df.groupby(['id']).mean().reset_index() #Groups data by id (assumed to be customer ID) and computes the average of all numeric columns per customer.
mean_6m = price_df[price_df['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index() #reset_index() keeps id as a regular column instead of an index.
mean_3m = price_df[price_df['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index() #Filters rows where price_date is after October 1, 2015 ... Gives average prices over the last 3 months
```

```python
#renaming and combining
def process_mean_data(df, prefix):
    df = df.rename(columns={
        "price_off_peak_var": f"{prefix}_price_off_peak_var",
        "price_peak_var": f"{prefix}_price_peak_var",
        "price_mid_peak_var": f"{prefix}_price_mid_peak_var",
        "price_off_peak_fix": f"{prefix}_price_off_peak_fix",
        "price_peak_fix": f"{prefix}_price_peak_fix",
        "price_mid_peak_fix": f"{prefix}_price_mid_peak_fix"
    })

    df[f"{prefix}_price_off_peak"] = df[f"{prefix}_price_off_peak_var"] + df[f"{prefix}_price_off_peak_fix"]
    df[f"{prefix}_price_peak"] = df[f"{prefix}_price_peak_var"] + df[f"{prefix}_price_peak_fix"]
    df[f"{prefix}_price_med_peak"] = df[f"{prefix}_price_mid_peak_var"] + df[f"{prefix}_price_mid_peak_fix"]

    return df
```

```python
mean_year = process_mean_data(mean_year, "mean_year")
mean_6m = process_mean_data(mean_6m, "mean_6m")
mean_3m = process_mean_data(mean_3m, "mean_3m")
```

```python
# Merge yearly and 6-month averages first
price_features = pd.merge(
    left=mean_year,
    right=mean_6m,
    on='id',
    how='inner'  # 'inner' ensures only customers present in both are kept
)

# Now merge with 3-month averages
price_features = pd.merge(
    left=price_features,
    right=mean_3m,
    on='id',
    how='inner'
)
```

```python
price_features.head()
```

#### Merging the price feature dataset with churn variable to check whether price sensitivity has any correlation with churn.

```python
# Merge price features with churn label
price_churn = pd.merge(price_features, client_df[['id', 'churn']], on='id')
```

```python
# Drop irrelevant columns before computing correlation
corr = price_churn.drop(columns=['id']).corr()

# Plot correlation matrix
plt.figure(figsize=(20, 18))
sns.heatmap(
    corr,
    cmap='Oranges',
    annot=True,
    fmt='.2f',            # Format annotations to 2 decimal places
    annot_kws={'size': 10},
    linewidths=0.5,       # Add subtle separation between cells
    square=True           # Make heatmap cells square-shaped
)

# Improve axis labels and ticks
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10, rotation=0)
plt.title('Correlation Between Price Features and Churn', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Create interactive heatmap
fig = px.imshow(
    corr,
    text_auto=True,               # Show correlation numbers
    aspect="auto",                # Auto-adjust cell size
    color_continuous_scale='Oranges',  # Use same color theme as before
    title='Correlation Matrix of Price Features and Churn'
)

# Improve layout
fig.update_layout(
    width=2000,
    height=2000,
    xaxis_title="Features",
    yaxis_title="Features",
    font=dict(size=10),
    title_x=0.5  # Center the title
)

# Optional: rotate x-axis labels for better readability
fig.update_xaxes(tickangle=45)
fig.update_yaxes(tickangle=0)

# Show the figure
fig.show()
```

 The above plot shows the correlation betweeen price variables with each other and with churn, however, the correlation between churn and prices variables is very low, which mean customers' churn is not sensitive to price change.
